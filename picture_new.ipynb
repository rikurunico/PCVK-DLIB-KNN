```python
import cv2
import dlib

# load the face detector and shape predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("models/shape_predictor_68_face_landmarks.dat")
```

```python
import os

# Import module yang digunkaan untuk ekstrak gambar
from preprocessing.input_to_image import get_frames_by_input_video
from preprocessing.scarpping_component import extract_component_by_images

pathDatasetImage = "dataset/video_to_images"
get_frames_by_input_video(pathInputVideo="dataset/video/EP02_01f.avi", pathOutputImage=pathDatasetImage, framePerSecond=60)

# looping semua file yang ada didalam
for filename in os.listdir(pathDatasetImage):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        # Create the directory if it doesn't exist
        image = cv2.imread(os.path.join(pathDatasetImage, filename))
        image = cv2.resize(image, (600, 500))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # detect the faces
        rects = detector(gray)

        # go through the face bounding boxes
        for rect in rects:
            # apply the shape predictor to the face ROI
            shape = predictor(gray, rect)

            # Buat variabel penambahan draw pixels
            pixel_shifting = {"pixel_x": 25, "pixel_y": 25}

            # Mata Kiri
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="eye_left", 
                objectStart=36, 
                objectEnd=42, 
                objectRectangle={"x_right": 39, "x_left": 36, "y_highest": 38, "y_lowest": 41}, 
                pixelShifting=pixel_shifting,
            )

            # Mata Kanan
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="eye_right", 
                objectStart=42, 
                objectEnd=48, 
                objectRectangle={"x_right": 45, "x_left": 42, "y_highest": 43, "y_lowest": 47}, 
                pixelShifting=pixel_shifting
            )

            # Alis Kiri
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="eyebrow_left", 
                objectStart=18, 
                objectEnd=22, 
                objectRectangle={"x_right": 21, "x_left": 17, "y_highest": 18, "y_lowest": 21}, 
                pixelShifting=pixel_shifting
            )

            # Alis Kanan
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="eyebrow_right", 
                objectStart=22, 
                objectEnd=27, 
                objectRectangle={"x_right": 26, "x_left": 22, "y_highest": 25, "y_lowest": 22}, 
                pixelShifting=pixel_shifting
            )

            # Mulut
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="mouth", 
                objectStart=46, 
                objectEnd=68, 
                objectRectangle={"x_right": 54, "x_left": 48, "y_highest": 52, "y_lowest": 57}, 
                pixelShifting=pixel_shifting
            )

            # Hidung Kanan
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="nose_right", 
                objectStart=31, 
                objectEnd=48, 
                objectRectangle={"x_right": 31, "x_left": 40, "y_highest": 40, "y_lowest": 48}, 
                pixelShifting={"pixel_x": 25, "pixel_y": -5}
            )

            # Hidung Kiri
            extract_component_by_images(
                image=image, 
                shape=shape, 
                frameName=filename.split(".")[0],
                objectName="nose_left", 
                objectStart=35, 
                objectEnd=54, 
                objectRectangle={"x_right": 47, "x_left": 35, "y_highest": 47, "y_lowest": 54}, 
                pixelShifting={"pixel_x": 25, "pixel_y": -5}
            )
            
            # tambah component disini

# cv2.imshow("Frame", image)
# cv2.waitKey(0)
```
